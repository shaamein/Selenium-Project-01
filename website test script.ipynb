```python
from selenium.webdriver import Chrome
from selenium.webdriver import ActionChains
import time
import random
import datetime

#shuffling function
def shufflelist (lists):
    
    random.shuffle(lists)
    
    return lists[0]

def letsgo(): # click lets go button
    buttons_id = browser.find_element_by_xpath('//*[@id="paragraph_button"]')
    buttons_id.click()
    return

def clicksubmit(): #click submit button
    buttons_id = browser.find_element_by_xpath('//*[@id="priority_answer"]')
    buttons_id.click()
    return

def clickbtn(path): #selecting and clicking an option
    
    buttons_id = browser.find_element_by_xpath(path)
    buttons_id.click()
    time.sleep(20)
    return


def draglist(question_num): #for iterating iDS
    letter = ['A', 'C', 'R', 'I', 'S', 'E']
    question_nums = [f'#Q{question_num}-{l}' for l in letter]

    return (question_nums)

def perform_drag(div1,div2,div3,q_list): #execute sort list
    
    for i in range(0,6):

        q_from = browser.find_element_by_css_selector(q_list[i])
        q_to = browser.find_element_by_xpath(f'/html/body/div[{div1}]/div[{div2}]/div[{div3}]/ul/li[{i+1}]')
    
        action_chains = ActionChains(browser)
        action_chains.drag_and_drop(q_from,q_to).perform()
    
    time.sleep(15)
    return

def answer_question(letter,question_num):
    num_of_btns = range(1,7) #returns a list of integers from 1 to 6
    btns = [f'//*[@id="{letter}-Q{question_num}-{btn}"]' for btn in num_of_btns]
    ret = shufflelist(btns)
    
    clickbtn(ret)
    return []


def id_checker(ids): #trial and error to determine the active category
    var = ids
    xpath_id = ''
    
    for letter in var:
        try:
            xpath = f'//*[@id="{letter}-Q1-1"]'
            this_id = browser.find_element_by_xpath(xpath)
            xpath_id = letter
            
        except:
            continue
            
        if xpath_id != '':
            return(xpath_id)
        
        else:
            print(f'error in {xpath_id}')
            
        
```

```python
##MAIN PROGRAM

browser = Chrome()
browser.get('##INSERT URL HERE##')
time.sleep(10)

begin_time = datetime.datetime.now()

#Q1
#when are you looking to attend college?
#button options
q1_btn1 = '//*[@id="btnThisYear"]'
q1_btn2 = '//*[@id="btn1to2Years"]'
q1_btn3 = '//*[@id="btnLookingToTransfer"]'
q1_list = [q1_btn1, q1_btn2, q1_btn3]
ret = shufflelist(q1_list)
# for checking
q1_dict = {'//*[@id="btnThisYear"]':'This Year',
           '//*[@id="btn1to2Years"]':'1 to 2 years',
           '//*[@id="btnLookingToTransfer"]':'Looking to Transfer'}
print (f'Q1: {q1_dict[ret]}')
clickbtn(ret)

#Q2
#what's your gender?
buttons = range (1,4)
q2_list = [f'//*[@id="wrap-ChatButton"]/button[{x}]' for x in buttons]
ret = shufflelist(q2_list)
# for checking
q2_dict = {'//*[@id="wrap-ChatButton"]/button[1]':'Male',
           '//*[@id="wrap-ChatButton"]/button[2]':'Female',
           '//*[@id="wrap-ChatButton"]/button[3]':'I choose not to identify'}
print (f'Q2: {q2_dict[ret]}')
clickbtn(ret)


#Q3
#Are you ready to take the HBCU matching assessment? Yes or No
#default is yes
q3_btn1 = '//*[@id="wrap-ChatButton"]/button[1]'
clickbtn(q3_btn1)
print ('Q3: Yes')


#Q4
#In the first question, please choose the answer that best describes you
#click lets go
letsgo()
time.sleep(20)
# I most like to...
q4_list = [f'//*[@id="wrap-chat"]/div[{x}]/label/input' for x in range(33,44,2)]
ret = shufflelist(q4_list)
# for checking
print ('Q4')
clickbtn(ret)


#Q5
# For the next 2 questions, sort the list of activities in order of
# preference. Move items up and down by dragging the icon on the right
letsgo()
time.sleep(15)

q5_list = draglist(1)
random.shuffle(q5_list)
# for checking
print ('Q5')

perform_drag(3,8,54,q5_list) ## numbers are use in xpath of Q7
clicksubmit()
time.sleep(10)


#Q6
#Next, imagine you work at a large hospital
#Sort the list in order of your preferred activities
letsgo()
time.sleep(15)

q6_list = draglist(2)
random.shuffle(q6_list)
# for checking
print ('Q6')

perform_drag(3,8,65,q6_list) ## numbers are use in xpath of Q7
clicksubmit()
time.sleep(10)

# You'll see a series of images and descriptions of things you might
# do at work. Tell us how much you like or dislike them
letsgo()
time.sleep(15)

categories = ['A', 'C', 'R', 'I', 'S', 'E']

for i in range(0,3):
    
    bins = categories
    pos = id_checker(bins)
    letters = pos
    for num in range(1,9):
        print(letters,num)
        answer_question(letters, num)
        
#results
buttons_id = browser.find_element_by_xpath('//*[@id="see_results_button"]')
buttons_id.click()

time.sleep(20)

print(datetime.datetime.now() - begin_time)


    
# //*[@id="FD_No"]
buttons_id = browser.find_element_by_xpath('//*[@id="FD_No"]')
buttons_id.click()
    
time.sleep(5)
print ('TEST SUCCESSFUL')
browser.close()
```
